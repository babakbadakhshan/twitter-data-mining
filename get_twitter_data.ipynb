```python
from ntscraper import Nitter
import pandas as pd
```

```python
scraper = Nitter(0)
```

```python
tweets = scraper.get_tweets('Ÿæÿßÿ±⁄©_ŸÑÿßŸÑŸá', mode='hashtag', number=2000)
```

```python
final_tweets = []
for x in tweets['tweets']:
    data = [x['link'], x['text'],x['date'],x['stats']['likes'],x['stats']['comments']]
    final_tweets.append(data)
final_tweets = pd.DataFrame(final_tweets, columns=['link', 'text', 'date', 'likes', 'comments'])
final_tweets
```

```python
final_tweets.to_csv('hormoz_tweets.csv')
```
